```python
def value_to_numeric(x):
    if type(x)==float or type(x)==int:
        return x
    
    if '%' in x:
        return float(x.replace('%',''))
       
    if ',' in x:
        return int(x.replace(',',''))
    
```

```python
df3['State_Share']=df3['State_Share'].apply(value_to_numeric)
```
